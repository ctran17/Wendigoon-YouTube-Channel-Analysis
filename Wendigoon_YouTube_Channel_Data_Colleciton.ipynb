```python
# Import YouTube API Key and Wendigoon YouTube channel ID
API_KEY = "AIzaSyDYUHcU11j_fmJVVfoD-BC6wNaGaeUJnbk"
CHANNEL_ID = "UC3cpN6gcJQqcCM6mxRUo_dA"

def get_video_ids(channel_id, api_key):
    # Initialize empty list to store video IDs
    video_ids =[]
    # Base URL for YouTube API search endpoint
    base_url = "https://www.googleapis.com/youtube/v3/search"
    params = {
        "part": "id",
        "channelId": channel_id,
        "maxResults": 50,
        "order": "date",
        "key": api_key
    }
    while True:
        response = requests.get(base_url, params = params)
        data = response.json()
        
        for item in data['items']:
            # Check if the item is a video
            if item['id']['kind'] == 'youtube#video':
                # If the item is video, append the video ID to the list
                video_ids.append(item['id']['videoId'])
                
        # Check if there is a next page token in the response
        if 'nextPageToken' in data:
            params['pageToken'] = data['nextPageToken']
        
        else:
            break
            
    # Return the list of video IDs
    return video_ids


def get_video_details(video_ids, api_key):
    # Initialize an empty list to store video details
    video_details = []
    # Base URL for YouTube API search endpoint
    base_url = "https://www.googleapis.com/youtube/v3/videos"
    for i in range(0, len(video_ids), 50):
        params = {
            "part": "snippet,statistics",
            "id": ','.join(video_ids[i:i+50]),
            "key": api_key
        }
        # Send the API request
        response = requests.get(base_url, params=params)
        # Convert the response to JSON
        data = response.json()
        
        for item in data['items']:
            # Create a dictionary to store video details
            video_data = {
                'title': item['snippet']['title'],   # Video title
                'date_created': item['snippet']['publishedAt'],   # Date & time created
                'view_count': item['statistics'].get('viewCount', 0),   # View count
                'comment_count': item['statistics'].get('commentCount', 0),   # Comment count
                'like_count': item['statistics'].get('likeCount', 0),   # Like count
            }
            video_details.append(video_data)
    
    return video_details

# Main execution
if __name__ == "__main__":
    video_ids = get_video_ids(CHANNEL_ID, API_KEY)
    video_details = get_video_details(video_ids, API_KEY)
    
    # Print the video details
    for video in video_details:
        print(video)
```

```python
# Transform the collected data into a CSV file for initial data cleaning and exploration
df = pd.DataFrame(video_details)
df.to_csv('wendigoon_channel_original_data.csv', index=False)

print("Wendigoon channel data collected and saved!")
```
